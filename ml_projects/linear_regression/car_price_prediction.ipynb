```
def find_univariable_linear_regression(
    dataframe, input_feature, output_feature, learning_rate, 
    theta_0 = 0, theta_1 = 0
    ):
    data_df = dataframe[[input_feature, output_feature]].copy()

    count_null_values = len(data_df) - len(data_df.dropna())
    if count_null_values > 0:
        print('There are', count_null_values, 'null values => drop them')
        data_df.dropna(inplace = True)
    m = len(data_df)
    print('Training set:', m)
    
    # Plot
    plt.plot(data_df[input_feature], data_df[output_feature], 'r.')
    plt.xlabel(input_feature)
    plt.ylabel(output_feature)
    plt.suptitle('Plot of input and output feature')
    plt.show()

    
    cost = calculate_cost(
        dataframe = data_df, 
        input_feature = input_feature, 
        output_feature = output_feature,
        theta_0 = theta_0,
        theta_1 = theta_1
    )
    print('Init theta_0, theta_1, cost: ', theta_0, theta_1, cost)


    cost_list = [cost]
    time_list = [0]

    t = 1
    is_continue = True
    while is_continue:
        derive_theta_0 = 0
        derive_theta_1 = 0
        for i in range(m):
            derive_theta_0 += theta_0 + theta_1*data_df[input_feature].iloc[i] - data_df[output_feature].iloc[i]
            derive_theta_1 += (theta_0 + theta_1*data_df[input_feature].iloc[i] - data_df[output_feature].iloc[i])*data_df[input_feature].iloc[i]
        derive_theta_0 /= m
        derive_theta_1 /= m

        theta_0 += - learning_rate*derive_theta_0
        theta_1 += - learning_rate*derive_theta_1

        cost = calculate_cost(
            dataframe = data_df, 
            input_feature = input_feature, 
            output_feature = output_feature,
            theta_0 = theta_0,
            theta_1 = theta_1
        )
        print('Iterate:', t, 
              ', new theta_0:', theta_0, 
              ', new theta_1:', theta_1, 
              ', new cost:', cost
        )

        cost_list.append(cost)
        time_list.append(t)
        t += 1

        if cost_list[-1] > cost_list[-2]:
            print('Cost is increasing') 
            break
        elif (cost_list[-1] - cost_list[-2])*100/cost_list[-2] > -0.01:
            is_continue = False
            # Ve do thi
            plt.plot(data_df[input_feature], data_df[output_feature], 'r.')
            plt.xlabel(input_feature)
            plt.ylabel(output_feature)

            x = np.linspace(0, 1000 ,1000)
            y = theta_0 + theta_1*x
            plt.plot(x, y, '-b', label='model')
            plt.legend(loc='upper left')
            plt.show()

            plt.plot(time_list, cost_list)
            plt.show()
            
        


def calculate_cost(
    dataframe, input_feature, output_feature, 
    theta_0, theta_1):
    m = len(dataframe)
    cost = 0
    for i in range(m):
        cost += (theta_0 + theta_1*dataframe[input_feature].iloc[i] - dataframe[output_feature].iloc[i])**2
    cost /= 2*m 
    return cost
```
